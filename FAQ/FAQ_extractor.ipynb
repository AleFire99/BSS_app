```python
import os
import pdfplumber
import json

# Folder containing the PDFs
folder_path = "QA"

# List all PDF files in the folder
pdf_files = [f for f in os.listdir(folder_path) if f.endswith(".pdf")]

# Prepare the tables data to be saved as JSON
all_table_data = []

# Iterate through each PDF file in the folder
for pdf_file in pdf_files:
    pdf_path = os.path.join(folder_path, pdf_file)
    
    # Open the PDF
    with pdfplumber.open(pdf_path) as pdf:
        tables = []
        
        # Extract tables from each page
        for page in pdf.pages:
            tables_on_page = page.extract_tables()
            tables.extend(tables_on_page)

        # Check if there are tables in the PDF
        if tables:
            # Get the first row of the first table to be used as the keys
            keys = tables[0][0]  # First row of the first table

            # Process the tables
            for table in tables:
                if len(table) > 1:
                    table_data = []
                    
                    # Iterate through the remaining rows and create a dictionary for each row
                    for row in table[1:]:
                        cleaned_row = [
                        cell.replace("\n", " ")
                        .replace("【", '"')
                        .replace("】", '"') if cell else "" 
                        for cell in row
                    ]
                        
                        # Ensure the cleaned_row matches the number of keys
                        if len(cleaned_row) == len(keys):
                            row_data = {keys[i]: cleaned_row[i] for i in range(len(keys))}
                            table_data.append(row_data)

                    # Append only the table content to the all_table_data list
                    all_table_data.extend(table_data)

# Output to a JSON file
output_file = "QA_sets.json"
with open(output_file, 'w', encoding='utf-8') as json_file:
    json.dump(all_table_data, json_file, ensure_ascii=False, indent=4)

print(f"All tables saved to {output_file}")
```

```python
import os
import pdfplumber
import json

# Folder containing the PDFs
folder_path = "QA"

# List all PDF files in the folder
pdf_files = [f for f in os.listdir(folder_path) if f.endswith(".pdf")]

# Prepare the tables data to be saved as JSON for tables with 3 columns
three_column_table_data = []

# Iterate through each PDF file in the folder
for pdf_file in pdf_files:
    pdf_path = os.path.join(folder_path, pdf_file)
    
    # Open the PDF
    with pdfplumber.open(pdf_path) as pdf:
        tables = []
        
        # Extract tables from each page
        for page in pdf.pages:
            tables_on_page = page.extract_tables()
            tables.extend(tables_on_page)

        # Process tables to handle 3-column tables
        keys = None  # To hold the keys (first row of the first table with 3 columns)
        
        for table in tables:
            # If the table has exactly 3 columns
            if len(table[0]) == 3:  # The first row should have 3 columns
                if keys is None:
                    # Use the first row of the first 3-column table as the keys
                    keys = table[0]
                table_data = []
                
                # Process the rows
                for row in table[1:]:
                    # Replace '\n' with ' ' and '【' and '】' with double quotes in each cell of the row
                    cleaned_row = [
                        cell.replace("\n", " ")
                        .replace("【", '"')
                        .replace("】", '"') if cell else "" 
                        for cell in row
                    ]
                    
                    # Ensure the cleaned_row matches the number of keys
                    if len(cleaned_row) == len(keys):
                        row_data = {keys[i]: cleaned_row[i] for i in range(len(keys))}
                        table_data.append(row_data)

                # Append the table data to the three_column_table_data list
                three_column_table_data.extend(table_data)

# Output to a JSON file
output_file = "QA_keywords.json"
with open(output_file, 'w', encoding='utf-8') as json_file:
    json.dump(three_column_table_data, json_file, ensure_ascii=False, indent=4)

print(f"All three-column tables saved to {output_file}")
```
