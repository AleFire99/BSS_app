```python
import os
import sqlite3
import requests
```

```python
# Create a folder to store images
if not os.path.exists("card_thumbnails"):
    os.makedirs("card_thumbnails")
```

```python
# Database file
DB_FILE = "C:/Users/AleFire/Desktop/Projects/BSS_app/telegram_bot/cards.db"  # Path inside the container
```

```python
def download_card_images(db_path):
    # Folder to save images
    base_folder = "card_thumbnails"
    os.makedirs(base_folder, exist_ok=True)

    # Connect to the database
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # Fetch all rows from the Cards table
    cursor.execute("SELECT CardID FROM Cards")
    rows = cursor.fetchall()

    for row in rows:
        card_id = row[0]
        subfolder_name = card_id.split("-")[0]  # Extract the part before '-'
        subfolder_path = os.path.join(base_folder, subfolder_name)
        os.makedirs(subfolder_path, exist_ok=True)

        # Build the URL and file path
        image_url = f"https://www.bssdb.dev/cards/bss/{card_id}.png"
        image_path = os.path.join(subfolder_path, f"{card_id}.png")

        try:
            # Download the image
            response = requests.get(image_url, stream=True)
            response.raise_for_status()

            # Save the image to the file system
            with open(image_path, "wb") as image_file:
                for chunk in response.iter_content(chunk_size=8192):
                    image_file.write(chunk)

            print(f"Downloaded {card_id} to {image_path}")
        except requests.exceptions.RequestException as e:
            print(f"Failed to download {card_id}: {e}")

    # Close the database connection
    conn.close()
```

```python
if __name__ == "__main__":
    download_card_images(DB_FILE)
```
